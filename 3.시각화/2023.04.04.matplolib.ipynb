#### Matplotlib
- 파이썬에서 사용되는 대표적인 데이터 시각화 라이브러리. 
- 2D, 3D 그래프, 히스토그램, 산점도, 파이차트 등 다양한 그래프를 지원하며, 
- 사용자가 그래프의 세부적인 설정을 조절할 수 있도록 다양한 옵션을 제공
- 핵심적인 개념은 Figure(그림)와 Axes(축)
- Figure는 그래프 전체를 의미하며, 하나의 Figure에 여러 개의 Axes를 추가하여 서로 다른 그래프를 그림 
- Axes는 Figure 내부의 그래프 영역을 의미하며, x축과 y축을 포함한 그래프 요소들을 생성하는 역할

np.linspace() 함수는 NumPy에서 제공하는 함수 중 하나로, 지정한 범위 내에서 균등한 간격으로 일정한 개수의 값을 생성하는 함수입니다.

np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)

start: 범위의 시작값

stop: 범위의 끝값

num: 생성할 값의 개수 (기본값: 50)

endpoint: 끝값을 포함할지 여부 (기본값: True)

retstep: 생성한 값의 간격(step)을 반환할지 여부 (기본값: False)

dtype: 생성한 값의 데이터 타입 (기본값: None)

axis: 배열의 축(axis) 방향 (기본값: 0)

```python
import numpy as np
import matplotlib .pyplot as plt
import pandas as pd
import seaborn as sns

# 데이터 생성

# x = [1, 2, 3, 4, 5]
# y = [2, 4, 6, 8, 10]

# x = np.arange(20)
# y = np.arange(20) * 2 + np.random.randn(20) *2

y = []
x = np.linspace(1, 10, 100) 
for i in x :
    r = i + np.random.normal(0,1)
    y.append(r)

plt.figure(figsize = (8, 5))

# 산정도 그래프 그리기

plt.scatter(x, y)

# 그래프 제목, x축 레이블, y축 레이블 설정

plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis', rotation = 360, labelpad = 20)

plt.show()
```

```python
import seaborn as sns

df = sns.load_dataset('titanic')
df.head()
```

```python
def get_age_cat(age):
    cat = ''
    
    if age < 15: cat = 'young'
    
    elif age < 30: cat = 'teen'
    
    elif age < 45: cat = 'adult'
    
    elif age < 60: cat = 'mature'
    
    else: cat = 'elder'
    
    return cat

df['age_cat'] = df['age'].apply(lambda x : get_age_cat(x))
df.head()
```

```python
df['Ratio'] = 1
df_age_cat = df.groupby('age_cat').sum()
df_age_cat
```

```python
df['Ratio'] = 1
df_age_cat = df.groupby('age_cat').sum()
df_age_cat.Ratio.plot(kind='pie',
                     figsize=(8,6),
                     autopct='%.2f%%',
                     startangle=0)
plt.title('Age Category', size=20)
plt.axis('equal') # 가로와 세로의 비율을 동일하게 설정
plt.legend(labels=df_age_cat.index,loc='best')
```

```python
print(df['fare'].describe())

plt.hist(df.fare, bins = 8)
```

```python
# 과제 fare 을 5개의 범주화 해서 fare_cat 컬럼을 생성하고 파이 차트로 시각화

df['fare'] = df['fare'].round(1)


df['fare_cat'] = pd.cut(df['fare'], [0, 8, 16, 40, 100, 600], labels = ['VL', 'L', 'M', 'H', 'VH'])

df_fare_cat = df.groupby('fare_cat').sum()
df_fare_cat.Ratio.plot(kind = 'pie', figsize = (10, 8), autopct = '%.2f%%', startangle= 0)

plt.title ('Fare_cat', size = 15)
plt.axis('equal')
plt.legend(labels = ['VL', 'L', 'M', 'H', 'VH'],loc='best')

df['fare_cat'].value_counts().sort_index()
```

```python
data['Wage'] = data['Wage'].str.replace('€', '', regex=False)
data['Wage'] = data['Wage'].str.replace('K', '', regex=False)

```

```python
data['Wage'] = data['Wage'].astype(int)

data['wage'] = pd.cut(data['Wage'], 100, labels = list(range(1, 101)))

plt.hist(data['wage'], bins = 100)
```

```python
import re

def get_num(x) :
    
    result = int(re.sub('[^0-9]', '', x))
    
    return result

data['nwage'] = data.Wage.apply(lambda x : get_num(x))

data[['Wage', 'nwage']].head()
```

```python
data.nwage.describe()
```

```python
data.nwage[data.nwage <= 100].plot(kind = 'hist', bins = 100)
```

```python
# 로그 변환 : np.log1p(x)

import numpy as np

def lognum(x) :
    
    result = np.log1p(x)
    
    return result

data['Lwage'] = data.nwage.apply(lambda x :lognum(x))

data[['Lwage', 'nwage']].head()
```
